## Binning and discretization

Binning is very important operation since many ML algorithms demand categorical attributes (e.g. association rules, some decision trees etc.). Additionally, binning (discretization) is used in exploratory analytics.

[pd.cut](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.cut.html) is convenient method that enables seamless bining and mapping of continuous features. Here we showcase simple example of binning and mapping, for details please refer to [pd.cut](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.cut.html)

```python
alcohol_category = pd.cut(wine['alcohol'], 5, labels=["very_low", "low", "medium", "high", "very_high"])
alcohol_category.value_counts()
```

```python
alcohol_category_map = {'very_low':'low', 'low':'low', 'medium':'medium', 'high':'high', 'very_high':'high'}
```

```python
alcohol_3_category =alcohol_category.apply(lambda x: alcohol_category_map[x])
```

```python
alcohol_3_category.value_counts()
```

# Missing Values
- Inacurate measurements
- Data does not exist (questionaries)

## Problems with missing values
- ML algorithms do not support missing values
- Can cause misleading statistics

## Solutions for missing values
- Removing rows
- Removing columns
- Imputation (replacing with known values)
- Do nothing (often usefull for descriptive analytics)

# What to do with Missing values
     - Drop rows
     - Drop columns
     - Impute

```python
ufo.dropna(how='any').shape # drop rows where any of columns have missing value / Also has inplace parameter
```

```python
ufo.dropna(how='all').shape
```

```python
ufo.dropna(subset=['City', 'Shape Reported'], how='any').shape
```

```python
ufo.dropna(subset=['City', 'Shape Reported'], how='any').shape # Drop if both attributes are null
```

```python
ufo['Shape Reported'].value_counts() # Missing values are excluded
```

```python
ufo['Shape Reported'].value_counts(dropna=False) # Missing values are excluded
```

```python
ufo['Shape Reported'].fillna(value='Various', inplace=True)
```
