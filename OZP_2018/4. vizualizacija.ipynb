# Distribution & Histogram

```python
sns.distplot(churn['CustServ Calls'])

#sns.distplot(churn['CustServ Calls'], bins=5)

sns.despine(offset=10, trim=True)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

# Box plot

```python
sns.boxplot(y=churn['CustServ Calls'])
sns.despine(offset=10, trim=True)
```

# Violin plot

```python
sns.violinplot(x=churn['Churn?'], y=churn['CustServ Calls'])
sns.despine(offset=10, trim=True)
```

```python
help(sns.violinplot)
```

# Pairplot

```python
sns.pairplot(churn.loc[:, 'Day Mins':'Intl Charge'])
sns.despine(offset=10, trim=True)
```

# Distribution

```python
h= sns.PairGrid(churn.loc[:, 'Day Mins':'Day Charge'])
h.map_offdiag(sns.kdeplot, cmap="Blues_d", n_levels=6)
sns.despine(offset=10, trim=True)
```
